```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Load the Excel file
file_path = 'codes comparison.xlsx'  
df = pd.read_excel(file_path)

# Assuming columns 'A' and 'B' contain the values in column A and B
column_A = df['codes']  
column_B = df['Jacksonville'] 

# Create a new column to track matches
df['Match'] = column_B.isin(column_A)

# Create a fill color for highlighting the matching cells
highlight_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# Load the workbook and select the active worksheet
wb = load_workbook(file_path)
ws = wb.active

# Highlight the matching cells in column B
for index, match in enumerate(df['Match']):
    if match:
        # Apply the highlight color to the corresponding cell in column B
        cell = ws[f'B{index + 2}']  # Adjust index for 1-based Excel indexing
        cell.fill = highlight_fill

# Save the highlighted version as a new Excel file
output_file = 'highlighted_comparison.xlsx'
wb.save(output_file)


df[['codes', 'Jacksonville', 'Match']].to_excel('comparison_results.xlsx', index=False)

print(f"Highlighted matches saved to: {output_file}")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Load the Excel file
file_path = 'codes comparison.xlsx'  
df = pd.read_excel(file_path)

# List of columns to compare with column A (starting from 'B' to 'G')
columns_to_compare = ['Jacksonville', 'Ocean_University', 'Morristown', 'Springfield', 'Williams_P', 'Baptist', 'Methodist Richordson M', 'Wisconson']


#  fill color for highlighting the matching cells
highlight_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# Load the workbook and select the active worksheet
wb = load_workbook(file_path)
ws = wb.active

# Dictionary to store similarity percentages
similarity_percentages = {}

# Iterate through each column in columns_to_compare and compare it with column A
for col in columns_to_compare:
    # Compare column A with each column
    df['Match'] = df[col].isin(df['codes'])
    
    # Calculate similarity percentage
    total_values = len(df[col])
    matched_values = df['Match'].sum()
    similarity_percentage = (matched_values / total_values) * 100
    similarity_percentages[col] = round(similarity_percentage, 2)
    
    # Highlight the matching cells in the respective column
    for index, match in enumerate(df['Match']):
        if match:
            # Apply the highlight color to the corresponding cell in the current column
            cell = ws[f'{chr(66 + columns_to_compare.index(col))}{index + 2}']  # Adjust for correct column letters (B-G)
            cell.fill = highlight_fill

# Save the highlighted version as a new Excel file
output_file = 'highlighted_comparison_multiple_columns.xlsx'
wb.save(output_file)


df[['codes'] + columns_to_compare + ['Match']].to_excel('comparison_results_multiple_columns.xlsx', index=False)

# Print similarity percentages in the console
for col, percentage in similarity_percentages.items():
    print(f"Similarity between 'codes' and '{col}': {percentage}%")

print(f"Highlighted matches saved to: {output_file}")
```

```python
### TO SHOW VIJAY
import pandas as pd
from openpyxl import load_workbook

# Load the highlighted Excel file
file_path = 'highlighted_comparison_multiple_columns.xlsx'  # Update this with your actual file path
wb = load_workbook(file_path)
ws = wb.active

# Load the Excel file into a DataFrame for easy manipulation
df = pd.read_excel(file_path)

# List of columns to process 
columns_to_compare = ['Jacksonville', 'Ocean_University', 'Morristown', 'Springfield', 'Williams_P', 'Baptist', 'Methodist Richordson M', 'Wisconson']

# Create an empty DataFrame to store non-highlighted values
non_highlighted_data = pd.DataFrame()

# Iterate through each column and extract non-highlighted cells
for col in columns_to_compare:
    non_highlighted_values = []
    
    # Iterate through each cell in the column (starting from the second row since first row is header)
    for row in range(2, ws.max_row + 1):
        cell = ws[f'{chr(66 + columns_to_compare.index(col))}{row}']  # Adjust column letter
        # Check if the cell is not highlighted (no fill or white fill)
        if cell.fill.start_color.index == '00000000' or cell.fill.start_color.index == 'FFFFFFFF':
            non_highlighted_values.append(cell.value)
    
    # Add the non-highlighted values to the DataFrame
    non_highlighted_data[col] = pd.Series(non_highlighted_values)

# Save the non-highlighted values to a separate Excel file named "non cpt codes.xlsx"
non_matching_output = 'non cpt codes.xlsx'
non_highlighted_data.to_excel(non_matching_output, index=False)

print(f"Non-highlighted values saved to: {non_matching_output}")
```
